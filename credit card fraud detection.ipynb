```python
x=df.drop(["Time","Class"],axis=1) #Splitting Data into x & y
y=df.Class
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x.head()
```

```python
#standardisation
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
#Splitting Data into Train & Test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=12)
```

```python
y.value_counts()
```

```python
#class imbalanced data
sns.countplot(x=y, palette="Set1")
```

```python
#Logistic regression 
from sklearn.linear_model import LogisticRegression
M1= LogisticRegression(max_iter=10000)    
M1.fit(x_train,y_train)
y_pred1=M1.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
#evaluation
acc1=accuracy_score(y_test,y_pred1)
print(acc1)
print(confusion_matrix(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```

```python
#undersasmpling
target="Class"
minority_class_len=len(df[df["Class"]==1])
print(minority_class_len)
```

```python
majority_class_indices=df[df["Class"]==0].index
print(majority_class_indices)
```

```python
random_majority_indices=np.random.choice(majority_class_indices,minority_class_len,replace=False)
print(len(random_majority_indices))
```

```python
minority_class_indices=df[df["Class"]==1].index
print(minority_class_indices)
```

```python
under_sample_indices=np.concatenate([minority_class_indices,random_majority_indices])
```

```python
#New Data after underSampling
under_sample=df.loc[under_sample_indices]
```

```python
sns.countplot(x=target,data=under_sample, palette="Set2")
```

```python
under_sample.head()
```

```python
#Creating new X and y data 
x=under_sample.drop(["Time","Class"],axis=1) 
y=under_sample.Class
```

```python
x_std= sc.fit_transform(x)
```

```python
# Splitting and creating new train & test data
P_train,P_test,q_train,q_test=train_test_split(x_std,y,test_size=0.33,random_state=12)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model= LogisticRegression(max_iter=10000)
model.fit(P_train,q_train)
q_pred1=model.predict(P_test)
```

```python
#model evaluation
acc6=accuracy_score(q_test,q_pred1)
print(acc6)
print(confusion_matrix(q_test,q_pred1))
print(classification_report(q_test,q_pred1))
```

```python
#SVM
from sklearn.svm import SVC
model2=SVC()
model2.fit(P_train,q_train)
q_pred2=model2.predict(P_test)
```

```python
#model evaluation
acc7=accuracy_score(q_test,q_pred2)
print(acc7)
print(confusion_matrix(q_test,q_pred2))
print(classification_report(q_test,q_pred2))
```

```python
#Random Forest 
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier(max_samples=0.75,random_state=12)
model3.fit(P_train,q_train)
q_pred3 = model3.predict(P_test)
```

```python
#model evaluation
acc8=accuracy_score(q_test,q_pred3)
print(acc8)
print(confusion_matrix(q_test,q_pred3))
print(classification_report(q_test,q_pred3))
```

```python
#gradient boosting
from sklearn.ensemble import GradientBoostingClassifier
model4 = GradientBoostingClassifier()
model4.fit(P_train,q_train)
q_pred4 = model4.predict(P_test)
```

```python
#model evaluation
acc9=accuracy_score(q_test,q_pred4)
print(acc9)
print(confusion_matrix(q_test,q_pred4))
print(classification_report(q_test,q_pred4))
```
